```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cereal_df = pd.read_csv("C:/Users/HP/Desktop/Data/DataAnalysis_raman/Classify spectrum/L-(-)-Tyrosine_allCutWaveNo.csv")
```

```python
# print(cereal_df)
X = cereal_df.to_numpy()
X = np.transpose(X)
wave_numbers = X[1, :]
intensities = X[2:, :]
print(intensities.shape)

for i in range(intensities.shape[0]):
  plt.plot(wave_numbers, intensities[i, :])
  plt.show()
```
