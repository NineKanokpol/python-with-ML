```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import rampy as rp
```

```python
Data=pd.read_table('Carbaryl (CB) -3_on foil_5x_Copy.txt')
X_CB1=Data[Data.columns[2]]
Y_CB1=Data[Data.columns[3]]
```

```python
X=[]
Y=[]
```

```python
for n in range(0,1011,1):
    X.append(X_CB1[n])
    Y.append(Y_CB1[n])
plt.plot(X_CB1,Y_CB1)
```

```python
X_arr = np.array(X)
Y_arr = np.array(Y)
print(X_arr.shape,Y_arr.shape)
```

```python
X_arr = X_arr.reshape(-1,1)
X_arr
```

```python
Y_arr = Y_arr.reshape(-1,1)
Y_arr
```

```python
for i in range(len(X_arr)):
    spec = X_arr[i]
spec
```

```python
for i in range(len(Y_arr)):
    spec1 = Y_arr[i]
spec1
```

```python
plt.plot(X_arr,Y_arr)
```

```python
roi = np.array([[600,800],[800,1000],[1000,1200],[1200,1400],[1400,1600]])
```

```python
for i in range(len(X_arr)):
    ycalc_als, base_als = rp.baseline(spec,spec1,roi,'als',lam=10**7,p=0.05)
```

```python
plt.plot(X_arr,ycalc_als)
```
