```python
plot.plot(wavenumber,X_all[0][:],label='D-tyrosine')
plot.plot(wavenumber,X_all[676][:],label='L-tyrosine')
plot.legend()
plot.xticks(fontsize = 15)
plot.yticks(fontsize = 15)
plot.xlabel( 'Wavanumber [cm-1]' ,fontsize = 15)
plot.ylabel( 'Intensity' ,fontsize = 15)
```

```python
plot.plot(wavenumber,X_all[0][:])
plot.xticks(fontsize = 15)
plot.yticks(fontsize = 15)
plot.xlabel( 'Wavanumber [cm-1]' ,fontsize = 15)
plot.ylabel( 'Intensity' ,fontsize = 15)
```

```python
spec = X_all[0][:]
x_min = 0
x_max = len(spec)
y_min = -1500 # min amplitude of spectrum
y_max = 29000 # max amplitude of spectrum

image_width = 128 # image width
image_height = 128 # image height

I = np.ones((image_height, image_width))
for i in np.arange(len(spec)):
    x_spec = i
    y_spec = spec[i]
     
    x_img = np.int(np.round((x_spec-x_min)/(x_max-x_min)*(image_width-1)))
    y_img = np.int(np.round((image_height-1) - (y_spec-y_min)/(y_max-y_min)*(image_height-1)), )
    I[y_img, x_img] = 0 
plot.figure(figsize = (7,7))
plot.imshow(I,cmap='gray')
plot.xticks(fontsize = 15)
plot.yticks(fontsize = 15)
plot.xlabel( 'Wavanumber [cm-1]' ,fontsize = 15)
plot.ylabel( 'Intensity' ,fontsize = 15)
plot.show()
```
